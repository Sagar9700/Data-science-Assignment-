```python
#33. Develop a program that generates a list of all possible combinations of two lists using list comprehension.

# initializing lists
list1 = [1, 3, 4]
list2 = [6, 7, 9]
list3 = [8, 10, 5]
 
# printing lists
print ("The original lists are : " + str(list1) +
                            " " + str(list2) +
                            " " + str(list3))
 
# using list comprehension
# to compute all possible permutations
res = [[i, j, k] for i in list1
                for j in list2
                for k in list3]
 
# printing result
print ("\nAll possible permutations are : " + str(res))
```

```python
#34. Write a program that calculates the mean, median, and mode of a list of numbers using list comprehension.
from collections import Counter

# List of numbers
numbers = [1, 2, 3, 4, 5, 5, 6, 6, 7, 8, 9, 9, 9]

# Calculate mean using list comprehension
mean = sum(numbers) / len(numbers)

# Calculate median using list comprehension
sorted_numbers = sorted(numbers)
n = len(sorted_numbers)
median = (sorted_numbers[n // 2] + sorted_numbers[(n - 1) // 2]) / 2 if n % 2 == 0 else sorted_numbers[n // 2]

# Calculate mode using list comprehension and Counter
counts = Counter(numbers)
max_count = max(counts.values())
mode = [num for num, count in counts.items() if count == max_count]

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```
